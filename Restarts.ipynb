```python
def generate_df(rest_df, one_alert_df):
    unique_alerts = []
    for index, row in rest_df.iterrows():
        if row['RESTART_DATETIME'].strftime("%H:%M") == "11:00" or row['RESTART_DATETIME'].strftime("%H:%M") == "10:00" or row['RESTART_DATETIME'].strftime("%H:%M") == "09:00" or row['RESTART_DATETIME'].strftime("%H:%M") == "08:00":
            rest_df.loc[index, 'SCHEDULED_RESTARTS'] = 1            
        for inner_ix, inner_row in one_alert_df.iterrows():
            time_diff =  time.mktime(row['RESTART_DATETIME'].timetuple()) - time.mktime(inner_row['OCCURED_DATETIME'].timetuple())
            if time_diff < 60*60*3 and time_diff > 0:
                rest_df.loc[index, one_alert_df.loc[inner_ix,'KEY'].strip()] = 1   
                unique_alerts.append(one_alert_df.loc[inner_ix,'KEY'].strip())
    rest_df = rest_df.fillna(0)        
    return rest_df, list(set(unique_alerts))

def calculate_conditional_probability(df, key):
    divider = len(df)
    if 'SCHEDULED_RESTARTS' in df.columns:
        divider = len(df.loc[df['SCHEDULED_RESTARTS'] == 0])
        
    if divider == 0:
        return 0, 0
    conditional_probability = len(df.loc[df[key] == 1])/divider
    unscheduled_restarts = divider/len(df)
    
    if conditional_probability > 1:
        conditional_probability = 1
        
    if unscheduled_restarts > 1:
        unscheduled_restarts = 1
        
    return conditional_probability, unscheduled_restarts
```

```python
prob_dict = {}
dfs_arr = []
for mac in restarts_df.MAC.unique():
    print(mac)
    res_df = restarts_df.loc[restarts_df['MAC'] == mac].copy()
    alert_df = time_alerts_df.loc[time_alerts_df['MAC'] == mac].copy()
    alert_df['OCCURED_DATETIME'] = pd.to_datetime(alert_df.OCCURED_TIME) 
    res_df['RESTART_DATETIME'] = pd.to_datetime(res_df.RESTARTS_TIME) 
    res_df = res_df[['RESTART_DATETIME']].copy()
    res_df['Restarts'] = np.repeat(1.0, len(res_df)).tolist()
    generated_df, unique_alerts = generate_df(res_df, alert_df)
    dfs_arr.append(generated_df.copy())
    print(unique_alerts)
    for unique_alert in unique_alerts:
        if unique_alert!= 'SCHEDULED_RESTARTS':
            con_prob, unsched_res_prob = calculate_conditional_probability(generated_df, unique_alert)
            if prob_dict.get(unique_alert) is not None: 
                prob_dict[unique_alert] = prob_dict[unique_alert]+ [con_prob]
            else:
                prob_dict[unique_alert] = [con_prob]
    print(prob_dict)
#     print(dfs_arr)
prob_dict            

# one_dev_alert_df = pd.to_datetime(one_dev_alert_df.ADDED_TIME) 

```
