```python
date_alerts = one_alerts['OCCURED_DATE'].value_counts().rename_axis('OCCURED_DATE').reset_index(name='alerts')
```

```python
x = pd.merge(date_restart, date_alerts, left_on='RESTART_DATE', right_on='OCCURED_DATE', how='left').drop('OCCURED_DATE', axis=1)
```

```python
x['date'] = pd.to_datetime(x["RESTART_DATE"])
x['day'] = x['date'].dt.dayofweek
```

```python
from fbprophet import Prophet
```

```python
x= x.rename(columns={'RESTART_DATE':'ds','restarts':'y'})
```

```python
x_model = Prophet(interval_width=0.95)
x_model.fit(x)
```

```python
x_forecast = x_model.make_future_dataframe(periods=36, freq='MS')
xforecast = x_model.predict(x_forecast)
```

```python
plt.figure(figsize=(18, 6))
x_model.plot(x_forecast, xlabel = 'Date', ylabel = 'Alerts')
plt.title('Alerts');
```
